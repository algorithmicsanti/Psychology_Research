```python
import numpy as np
import matplotlib.pyplot as plt

frecuencia = 15  
frecuencia_de_muestreo = 250 
duracion = 2
t = np.linspace(0, duracion, duracion * frecuencia_de_muestreo)  
x = np.sin((2 * np.pi) * frecuencia *t) 

# Generacin de ruido aleatorio (ruido con distribucin normal)
ruido_g = np.random.randn(len(t)) * 0.5  
ruido_uniforme = np.random.uniform(-0.5, 0.5, len(t))

# Seal ruidosa (seal original + ruido)
xuniforme = x + ruido_uniforme
xg = x + ruido_g


# Grfica
plt.figure(figsize=(10, 5))
plt.plot(t, xuniforme)
plt.title(f"Sine Wave: Frequency {frecuencia} Hz, Duration {duracion} s, Frecuencia de muestreo {frecuencia_de_muestreo} points/s")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.grid(True)

plt.show()

plt.figure(figsize=(10, 5))
plt.plot(t, xg)
plt.title(f"Sine Wave: Frequency {frecuencia} Hz, Duration {duracion} s, Frecuencia de muestreo {frecuencia_de_muestreo} points/s")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.grid(True)

plt.show()
```

```python
# Definir los parmetros
x = np.linspace(0, 5, 500)  # Rango de x de 0 a 5

# Funcin exponencial
y = np.exp(x)

# Generar ruido gaussiano y ruido uniforme
ruido_gaussiano = np.random.randn(len(x)) * 0.5  # Ruido gaussiano (escalado)
ruido_uniforme = np.random.uniform(-5, 5, len(x))  # Ruido uniforme (rango [-0.2, 0.2])

# Seales con ruido
y_gaussiana = y + ruido_gaussiano
y_uniforme = y + ruido_uniforme

# Crear dos grficas
plt.figure(figsize=(12, 6))

# Grfica 1: Funcin exponencial con ruido gaussiano
plt.subplot(1, 2, 1)  # Subgrfica 1 (1 fila, 2 columnas, 1ra grfica)
plt.plot(x, y, label="Exponencial pura", linestyle='--')
plt.plot(x, y_gaussiana, label="Con ruido gaussiano")
plt.title("Funcin exponencial con ruido gaussiano")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

# Grfica 2: Funcin exponencial con ruido uniforme
plt.subplot(1, 2, 2)  # Subgrfica 2 (1 fila, 2 columnas, 2da grfica)
plt.plot(x, y, label="Exponencial pura", linestyle='--')
plt.plot(x, y_uniforme, label="Con ruido uniforme")
plt.title("Funcin exponencial con ruido uniforme")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

# Mostrar ambas grficas
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros
fo = 15  # Frecuencia 15 Hz
delta_x = 1 / 250  # delta x es 1/250
m = 1  # Factor multiplicativo (asumido 1 para simplificacin)
t = np.linspace(0, 2, 2 * 250)  # Tiempo de 2 segundos con 250 puntos por segundo

# Ecuacin de la onda: sen(2 * m * fo * delta_x * t)
x = np.sin(2 * np.pi * fo * delta_x * t)

# Graficando la onda
plt.figure(figsize=(10, 5))
plt.plot(t, x)
plt.title("Onda Cerebral: Frecuencia 15 Hz, Muestreo 250 puntos/s")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
```
