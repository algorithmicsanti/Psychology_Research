```python
# Definir los parámetros
x = np.linspace(0, 5, 500)  # Rango de x de 0 a 5

# Función exponencial
y = np.exp(x)

# Generar ruido gaussiano y ruido uniforme
ruido_gaussiano = np.random.randn(len(x)) * 0.5  # Ruido gaussiano (escalado)
ruido_uniforme = np.random.uniform(-5, 5, len(x))  # Ruido uniforme (rango [-0.2, 0.2])

# Señales con ruido
y_gaussiana = y + ruido_gaussiano
y_uniforme = y + ruido_uniforme

# Crear dos gráficas
plt.figure(figsize=(12, 6))

# Gráfica 1: Función exponencial con ruido gaussiano
plt.subplot(1, 2, 1)  # Subgráfica 1 (1 fila, 2 columnas, 1ra gráfica)
plt.plot(x, y, label="Exponencial pura", linestyle='--')
plt.plot(x, y_gaussiana, label="Con ruido gaussiano")
plt.title("Función exponencial con ruido gaussiano")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

# Gráfica 2: Función exponencial con ruido uniforme
plt.subplot(1, 2, 2)  # Subgráfica 2 (1 fila, 2 columnas, 2da gráfica)
plt.plot(x, y, label="Exponencial pura", linestyle='--')
plt.plot(x, y_uniforme, label="Con ruido uniforme")
plt.title("Función exponencial con ruido uniforme")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

# Mostrar ambas gráficas
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
fo = 15  # Frecuencia 15 Hz
delta_x = 1 / 250  # delta x es 1/250
m = 1  # Factor multiplicativo (asumido 1 para simplificación)
t = np.linspace(0, 2, 2 * 250)  # Tiempo de 2 segundos con 250 puntos por segundo

# Ecuación de la onda: sen(2 * m * fo * delta_x * t)
x = np.sin(2 * np.pi * fo * delta_x * t)

# Graficando la onda
plt.figure(figsize=(10, 5))
plt.plot(t, x)
plt.title("Onda Cerebral: Frecuencia 15 Hz, Muestreo 250 puntos/s")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
```
