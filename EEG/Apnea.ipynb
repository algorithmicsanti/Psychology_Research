```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import matplotlib.pyplot as plt
df_1 = pd.read_csv('C:/Users/santi/Downloads/apnea.txt', delim_whitespace=True)
```

```python
df_1
```

```python

# Crear la figura y los ejes
plt.figure(figsize=(10, 6))

# Graficar los tres canales usando .iloc para referirse a las columnas por su posiciÃ³n
plt.plot(df_1.index, df_1.iloc[:, 0], label='Canal 1')  # Primera columna: Canal 1
plt.plot(df_1.index, df_1.iloc[:, 1], label='Canal 2')  # Segunda columna: Canal 2
plt.plot(df_1.index, df_1.iloc[:, 2], label='Canal 3')  # Tercera columna: Canal 3

# Etiquetas y tÃ­tulo
plt.title('GrÃ¡fica de Apnea: Voltaje vs Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Voltaje (V)')

# Mostrar la leyenda
plt.legend()

# Mostrar la grÃ¡fica
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Derivada de cada canal
derivada_canal_1 = np.gradient(df_1.iloc[:, 0])
derivada_canal_2 = np.gradient(df_1.iloc[:, 1])
derivada_canal_3 = np.gradient(df_1.iloc[:, 2])

plt.figure(figsize=(10, 6))

# Graficar las derivadas
plt.plot(df_1.index, derivada_canal_1, label='Derivada Canal 1')
plt.plot(df_1.index, derivada_canal_2, label='Derivada Canal 2')
plt.plot(df_1.index, derivada_canal_3, label='Derivada Canal 3')

plt.title('Derivada de la SeÃ±al: Voltaje vs Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Derivada del Voltaje (V/s)')

# Mostrar la leyenda
plt.legend()

# Mostrar la grÃ¡fica
plt.grid(True)
plt.show()

```
