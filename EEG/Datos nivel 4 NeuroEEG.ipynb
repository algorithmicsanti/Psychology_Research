```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler


df = pd.read_csv('C:/Users/santi/Downloads/NIVEL4.txt', delim_whitespace=True)
df = df.drop([0, 1]).reset_index(drop=True)
df

```

```python
scaler_standard = StandardScaler()
df_standard_scaled = pd.DataFrame(scaler_standard.fit_transform(df), columns=df.columns)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(8, 5))
for i in range(2):  
    plt.plot(df_standard_scaled.iloc[:1000, i], label=f'Canal {i+1}')

plt.title('Datos EEG Normalizados (Z-score)')
plt.xlabel('Muestras')
plt.ylabel('Amplitud Normalizada')
plt.legend()
plt.grid(True)
plt.show()

```
