```python
import pandas as pd
import numpy as np
```

```python
# Cargar el archivo
archivo_eeg = r'C:/Users/santi/Downloads/JIMBAI25-04.txt'
datos_eeg = pd.read_csv(archivo_eeg, sep='\s+', header=None)

# Dividir los datos en tres fases
fase1 = datos_eeg.iloc[:300000, :]  # Primer segmento
fase2 = datos_eeg.iloc[300000:350000, :]  # Segundo segmento
fase3 = datos_eeg.iloc[350000:, :]  # Tercer segmento

# Calcular promedio y desviación estándar para cada canal en cada fase
estadisticas_fase1 = pd.DataFrame({
    'Promedio': fase1.mean(),
    'Desviación Estándar': fase1.std()
})

estadisticas_fase2 = pd.DataFrame({
    'Promedio': fase2.mean(),
    'Desviación Estándar': fase2.std()
})

estadisticas_fase3 = pd.DataFrame({
    'Promedio': fase3.mean(),
    'Desviación Estándar': fase3.std()
})

# Concatenar las estadísticas en una tabla final
estadisticas_totales = pd.concat([estadisticas_fase1, estadisticas_fase2, estadisticas_fase3], 
                                 keys=['Fase 1', 'Fase 2', 'Fase 3'], axis=1)

# Mostrar la tabla final
print(estadisticas_totales)

# Guardar la tabla como CSV si es necesario
estadisticas_totales.to_csv('estadisticas_fases_eeg.csv')


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo
archivo_eeg = r'C:/Users/santi/Downloads/JIMBAI25-04.txt'
datos_eeg = pd.read_csv(archivo_eeg, sep='\s+', header=None)

# Dividir los datos en tres fases
fase1 = datos_eeg.iloc[:300000, :]  # Primer segmento
fase2 = datos_eeg.iloc[300000:303500, :]  # Segundo segmento
fase3 = datos_eeg.iloc[303500:, :]  # Tercer segmento

# Calcular el promedio para cada canal en cada fase
promedios_fase1 = fase1.mean()
promedios_fase2 = fase2.mean()
promedios_fase3 = fase3.mean()

# Crear scatter plot combinado
plt.figure(figsize=(10, 6))

# Fase 1
plt.scatter(range(1, len(promedios_fase1) + 1), promedios_fase1, color='blue', label='Fase 1')

# Fase 2
plt.scatter(range(1, len(promedios_fase2) + 1), promedios_fase2, color='green', label='Fase 2')

# Fase 3
plt.scatter(range(1, len(promedios_fase3) + 1), promedios_fase3, color='red', label='Fase 3')

# Configurar el gráfico
plt.title('Promedio del voltaje por canal para cada fase')
plt.xlabel('Canales')
plt.ylabel('Voltaje promedio (uV)')
plt.xticks(range(1, len(promedios_fase1) + 1))  # Etiquetas de los canales
plt.grid(True)

# Mostrar leyenda
plt.legend()

# Mostrar gráfico
plt.show()
```
